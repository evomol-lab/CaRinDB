#3 - Creating the attribute in the same format as the AlphaMissense *protein_variant* attribute: *changeProt_am*.


```
df_33.info()
```

```
df_33[["changeProt_EFF", "aminBefore", "aminAfter", "poschangeProt"]]
```

```
def get_Before(aa):
  aa_mapping = {'Ala': 'A', 'Arg': 'R', 'Asn': 'N', 'Asp': 'D', 'Cys': 'C',
                   'Gln': 'Q', 'Glu': 'E', 'Gly': 'G', 'His': 'H', 'Ile': 'I',
                   'Leu': 'L', 'Lys': 'K', 'Met': 'M', 'Phe': 'F', 'Pro': 'P',
                   'Ser': 'S', 'Thr': 'T', 'Trp': 'W', 'Tyr': 'Y','Val':'V'}
  return aa_mapping.get(aa)
```

```
def get_After(aa):
  aa_mapping = {'Ala': 'A', 'Arg': 'R', 'Asn': 'N', 'Asp': 'D', 'Cys': 'C',
                   'Gln': 'Q', 'Glu': 'E', 'Gly': 'G', 'His': 'H', 'Ile': 'I',
                   'Leu': 'L', 'Lys': 'K', 'Met': 'M', 'Phe': 'F', 'Pro': 'P',
                   'Ser': 'S', 'Thr': 'T', 'Trp': 'W', 'Tyr': 'Y','Val':'V'}
  if aa in ['Phe','His','Ile', 'Lys','Leu','Met','Thr','Trp','Val', 'Asp','Glu','Ala','Arg','Asn','Cys','Gly','Gln','Pro','Ser','Tyr' ]:
    return aa_mapping.get(aa)
  else: #valores: *, ., Ter
    return aa
```

```
#Creating the attribute
df_33["changeProt_am"] = df_33["aminBefore"].apply(get_Before) + df_33["poschangeProt"].astype(str) + df_33["aminAfter"].apply(get_After)
```

```
df_33[["changeProt_EFF", "aminBefore", "aminAfter", "poschangeProt", "changeProt_am"]]
```

```
df_33.info()
```

#4 - Generating attributes derived from the AlphaMissense database

The database with the **33 cancers** will be joined with the **AlphaMissense** database through the attributes: **uniprot_id** and **changeProt_am**. And it will add the following attributes to this database:
- **am_pathogenicity**:AlphaMissense Score
-	**am_class**: Identifies if the mutation is **ambiguous**, **benig** or **pathogenic**.

##4.1 - Joining the  base_33 table (through the fields *Uniprot_id* and *changeProt_am*) with the  AlphaMissense Table (through the fields *uniprot_id* and *protein_variant*) to obtain the attributes: *am_pathogenicity* and  *am_class*

```
import pandas as pd
base_merge = pd.merge(df_33, base_AM, left_on=['Uniprot_id', 'changeProt_am'], right_on=['uniprot_id','protein_variant'], how='left')
```

```
base_merge.info(max_cols=120)
```

```
base_Nan = base_merge[base_merge.isna().any(axis=1)]
```
