```
base_PDB_wild.info()
```

```
base_PDB_wild.head()
```

```
#Identify duplicates records in the data
dupes=base_PDB_wild.duplicated()
sum(dupes)
```

```
#removing duplicates
base_PDB_wild_ok = base_PDB_wild.drop_duplicates(keep='first')
```

```
#Identify duplicates records in the data
dupes=base_PDB_wild_ok.duplicated()
sum(dupes)
```

```
base_PDB_wild_ok.info()
```

```
#Put the PDBs in numerical order
df_sort = base_PDB_wild_ok.sort_values(by=['PDB_wild_id'], ascending=True)
```
