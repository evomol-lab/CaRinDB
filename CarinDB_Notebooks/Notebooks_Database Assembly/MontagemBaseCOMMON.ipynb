#2 - Processing of COMMON databases

From the file **00-All.vcf.gz** available at https://ftp.ncbi.nih.gov/snp/organisms/human_9606/VCF/ the fields were extracted: **Chrom**, **Pos**, **SNP_ID**, **REF**, **ALT**, **COMMON** and the file **00-All.tsv** was generated. This was split into 7 files, due to its size of 20 GB (using the Linux command: $ split -b3000000000 00-All.tsv).

###2.1.1 Treatment of the **REF** field

The database will be inspected to check if there are instances of the **REF** attribute that have multivalued values, separated by **','**. These instances need to be treated to have atomic values.

```
df_common1 = pd.read_csv("drive/My Drive/BaseNovaDaniRaul/Common_hg38_1.csv",delimiter='\t')
```

```
df_common1['REF']
```

```
# Searching for instances that have more than one value for REF
# Pattern to be searched: A,T,C,G
df_common1_mult = df_common1.query('REF.str.contains(",")', engine='python')
```

```
df_common1_mult.info()
```

The **REF** field in the **COMMON_01** database is not a multivalued field, so it does not need treatment

###2.1.2 Treatment of the *ALT* field

The database will be inspected to check if there are instances of the **ALT** attribute that have multivalued values, separated by **','**. These instances need to be treated to have atomic values.

```
df_common1['ALT']
```

```
# Searching for instances that have more than one value for ALT
# Pattern to be searched: A,T,C,G
df_common1_mult = df_common1.query('ALT.str.contains(",")', engine='python')
```

```
df_common1_mult.info()
```

The **REF** field in the COMMON_07 database is not a multivalued field, so it does not need treatment.

###2.7.2 Treatment of the *ALT* field

The database will be inspected to check if there are instances of the **ALT** attribute that have multivalued values, separated by **','**. These instances need to be treated to have atomic values.

```
df_common7['ALT']
```

```
# Searching for instances that have more than one value for ALT
# Pattern to be searched: A,T,C,G
df_common7_mult = df_common7.query('ALT.str.contains(",")', engine='python')
```

```
df_common7_mult.info()
```

```
#Deleting from df_common7 the instances that have more than one value for ALT
df_common7 = df_common7[~df_common7.ALT.str.contains(",")]
```

```
df_common7.info()
```

###2.7.3 -  Generating two files with the content of the *df_common7* database without NaN values

Two files will be generated:

- common_hg38_7.csv

- common_hg38_mult_7.csv (the records that have the multivalued ALT field.)

```
df_common7.to_csv("drive/My Drive/BaseNovaDaniRaul/Common_hg38_7.csv",sep='\t',index=False)
df_common7_mult.to_csv("drive/My Drive/BaseNovaDaniRaul/Common_hg38_mult_7.csv",sep='\t',index=False)
```
